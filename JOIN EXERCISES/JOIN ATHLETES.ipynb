```python
filepath = "distributed-data/"

data = pd.read_csv("distributed-data/001.csv")
final_length = len(data)
#import different csv, and append to the last one 
for i in range(2,333): #From second file to 332
    if i < 10:
        filename = "00" + str(i) 
    if 10 <= i < 100:
        filename = "0" + str(i)
    #if i ==35: If i have problems with missing csv 
        #pass
    if i >= 100:
        filename = str(i)
    file = filepath + filename + ".csv"
    
    temp_data = pd.read_csv(file)
    final_length += len(temp_data)
    
    data = pd.concat([data, temp_data], axis = 0)
```
