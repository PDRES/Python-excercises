```python
class Vehicles():
    
    def __init__(self, color, wheels):
        self.color = color
        self.wheels = wheels
        
    def __str__(self):
        return "color {}, {} wheels".format( self.color, self.wheels )
        
        
class Car(Vehicles):
    
    def __init__(self, color, wheels, speed, cilinders, power ):
        super().__init__(color, wheels)  
        self.speed = speed
        self.cilinders = cilinders
        self.power=power
        
    def __str__(self):
        return super().__str__() + ", {} km/h, {} cc, {} CV ".format(self.speed, self.cilinders, self.power)

class Truck(Car):
    
    def __init__(self, color, wheels, speed, cilinders, power, load):
        super().__init__(color, wheels,speed,cilinders,power)
        self.load=load
        
    def __str__(self):
        return super().__str__() + ", {}Tons".format(self.load)


class Bycicle(Vehicles):
    
    def __init__(self, color, wheels,typee):
        super().__init__(color, wheels)  
        self.typee=typee
        
    def __str__(self):
        return super().__str__() + ", style {}".format(self.typee)
        
class Motorbike(Bycicle):
    
    def __init__(self, color, wheels, typee, cilinders,speed):
        super().__init__(color, wheels,typee)  
        self.cilinders=cilinders
        self.speed=speed
        
    def __str__(self):
        return super().__str__() + ", {} cilinders, and {} km/h".format(self.cilinders, self.speed)
    
    

    
def cataloge(Vehicles, wheels=None):
        
    if wheels != None:
        count = 0
        for v in Vehicles:
            if v.wheels == wheels: 
                count += 1
        print("\nfound {} vehicles with {} wheels:".format(count, wheels))

    for v in Vehicles:
        if wheels == None:
            print("{}{}".format(type(v).__name__, v))
        else:
            if v.wheels == wheels:
                print("{} {}".format(type(v).__name__, v))

        
listt = [
    Car("blue", 4, 150, 1200,300),
    Truck("white", 8, 100, 1300, 1500,2000),
    Bycicle("green", 2, "urban"),
    Motorbike("black", 2, "sport", 2, 900)]

cataloge(listt)
cataloge(listt, 2)
cataloge(listt, 8)
cataloge(listt, 4)
cataloge(listt, 3)
```
